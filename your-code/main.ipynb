#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
month_encoded = fitbit.groupby(["Months_encoded"])["Steps"].median().reset_index()
month_encoded
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
week_weekend = fitbit.groupby(["Work_or_Weekend"])["Steps"].median().reset_index()
week_weekend

x = month_encoded["Months_encoded"]
y = month_encoded["Steps"]

x2= week_weekend["Work_or_Weekend"]
y2 = week_weekend["Steps"]

f, [month, work] = plt.subplots(1,2, figsize = (10,8))

plt.sca(month)
plt.scatter(x, y)
plt.title("Steps month")

plt.sca(work)
plt.bar(x2, y2)
plt.title("Steps by work or weekend")

f.savefig('../figures/steps.png')
```

#### Plot the correlation matrix using seaborn.

```python
# your code here
#sns.pairplot(titanic)

corr = titanic.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=np.bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
